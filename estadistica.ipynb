```python
import numpy as np
import pandas as pd
import matplotlib as mlp
import matplotlib.pyplot as mlp
import seaborn as sns
```

```python
data_player = pd.read_csv('./Datos/player_valuations.csv')
data_player.head()
# data_player
```

```python
#Cambiamos el nombre de la columna de inglés a español
data_player.rename(columns={'market_value_in_eur':'precio_en_euros'}, inplace=True)
data_player.rename(columns={'player_id':'id_jugador'}, inplace=True)
data_player.rename(columns={'date':'fecha'}, inplace=True)
data_player.rename(columns={'current_club_id':'id_club_actual'}, inplace=True)
data_player.rename(columns={'jugador_club_competicion_nacional_id':'competicion_nacional_id'}, inplace=True)
data_player.head()
```

```python
#Ordenar el DataFrame en función de la columna precio_en_euros de manera descendente
df_sorted = data_player.sort_values(by='precio_en_euros', ascending=False)
df_sorted
```

1. Resúmen Estadístico Básico:

```python
#Configurar el formato de visualización para mostrar números completos
pd.set_option('display.float_format', lambda x: '%.3f' %x)
df_sorted.describe()
```

```python
#Funciones de Pandas
#Media
df_sorted['precio_en_euros'].mean()
```
